```python
import pandas as pd #light package to do multiple array for 2 dimension
import numpy as np #package for data frame
import matplotlib.pyplot as plt #do some visualisation
import seaborn as sns #to create some sort of histogram or heap map
```

```python
df = pd.read_csv('supermarket_sales.csv')
```

```python
plt.title('Comparision of Gender Among All Branches')
plt.pie(df.Gender.value_counts(), explode=(0.025,0.025), labels=df.Gender.value_counts().index, autopct='%1.1f%%')
plt.legend()
plt.show()
```

```python
branch_totals=df.groupby('Branch')['Total'].sum().reset_index()
```

```python
plt.title('Total Sales of Each Branch')
plt.bar(branch_totals['Branch'], branch_totals['Total'] )
plt.xlabel('Branch')
plt.ylabel('Total Sales')
plt.legend()
plt.show()
```

```python
payment_counts = df['Payment'].value_counts()
```

```python
plt.title('Most Used Payment Method')
plt.bar(payment_counts.index, payment_counts)
plt.xlabel('Payment Method')
plt.ylabel('Amount of Times Used')
plt.legend()
plt.show()
```

```python
df['Customer type'].value_counts().plot(kind='pie', autopct='%1.2f%%', colors=['skyblue', 'lightgreen'])
plt.title('Distribution of Customer Types')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Product line', hue='Gender', data=df)
plt.title('Number of Each Gender by Product line')
plt.xlabel('Product line')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=df['Product line'], y=df['gross income'], palette='hot')
plt.title('Gross Income by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Gross Income')

plt.show()
```
